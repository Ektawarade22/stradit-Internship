```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)

# Data Cleaning

df["Age"] = df["Age"].fillna(df["Age"].median())

df["Embarked"] = df["Embarked"].fillna(
    df["Embarked"].mode()[0]
)

# Feature Engineering

df["FamilySize"] = (
    df["SibSp"] + df["Parch"] + 1
)

df["IsAlone"] = (
    df["FamilySize"] == 1
).astype(int)

# Create Age Groups

bins = range(0, 85, 5)

df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=bins
)

# FIXED WARNING HERE
avg_fare = df.groupby(
    "AgeGroup",
    observed=False
)["Fare"].mean()

# Create Dashboard
fig, axes = plt.subplots(
    1,
    3,
    figsize=(18, 5)
)

# Chart 1 — Survival Rate by Sex
sex_rate = df.groupby("Sex")["Survived"].mean()

axes[0].bar(
    sex_rate.index,
    sex_rate.values
)

axes[0].set_title(
    "Survival Rate by Sex"
)

axes[0].set_xlabel("Sex")

axes[0].set_ylabel(
    "Survival Rate"
)

# Chart 2 — Age Distribution
axes[1].hist(
    df["Age"],
    bins=20
)

axes[1].set_title(
    "Age Distribution"
)

axes[1].set_xlabel("Age")

axes[1].set_ylabel(
    "Frequency"
)

# Chart 3 — Average Fare by Age Group
axes[2].plot(
    avg_fare.index.astype(str),
    avg_fare.values,
    marker="o"
)

axes[2].set_title(
    "Average Fare by Age Group"
)

axes[2].set_xlabel(
    "Age Group"
)

axes[2].set_ylabel(
    "Average Fare"
)

axes[2].tick_params(
    axis="x",
    rotation=90
)

# Layout + Save

plt.tight_layout()

plt.savefig(
    "week2_deliverable.png"
)

plt.show()

# -----------------------------
# Summary
# -----------------------------

print(
    "Females had a much higher survival rate than males."
)

print(
    "Most passengers were between 20 and 40 years old."
)

print(
    "Fare distribution was heavily right-skewed."
)

print(
    "Passengers in higher classes generally paid higher fares."
)

print(
    "Passengers travelling alone had lower survival chances."
)
```
