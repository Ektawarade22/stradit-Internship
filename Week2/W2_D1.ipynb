```python
#task 4
def remove_all(lst, value):
    new_list = []
    for item in lst:
        if item != value:
            new_list.append(item)
    return new_list
numbers = [1, 2, 3, 4, 5, 2, 4, 1, 5]
result = remove_all(numbers, 2)
print(result)
```

```python
#task 3
def is_pangram(sentence):
    letters=set("abcdefghtijklmnopqrstuvwxyz")
    return set(sentence.lower()).issuperset(letters)
text=input("Enter sentence:")
print(is_pangram(text))
```

```python
#Block D=Dictionaries
#task 1
sent=input("Enter a sentence:")
words=sent.split()
freq={}
for word in words:
    if word in freq:
        freq[word]+=1
    else:
        freq[word]=1
print(freq)
```

```python
#task 5
transactions=[
    {"user":"a","amt":10},
    {"user":"b","amt":15},
    {"user":"c","amt":20}
]
total={}
for t in transactions:
    user=t["user"]
    amt=t["amt"]

    if user in total:
        total[user]+=amt
    else:
        total[user]=amt
print(total)
```
