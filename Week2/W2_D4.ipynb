```python
import pandas as pd

# =========================
# Load Titanic Dataset
# =========================

df = pd.read_csv("titanic.csv")


# =========================
# Cleaning Steps
# =========================

# Fill missing Age with median age
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill missing Embarked with most common value
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Drop Cabin column
df = df.drop(columns=["Cabin"])


# =========================
# Feature Engineering
# =========================

# Extract Title from Name
df["Title"] = (
    df["Name"]
    .str.split(", ").str[1]
    .str.split(".").str[0]
)

# Create FamilySize column
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

# Create IsAlone column
df["IsAlone"] = (df["FamilySize"] == 1).astype(int)

# Create FareCategory column
df["FareCategory"] = pd.cut(
    df["Fare"],
    bins=[0, 10, 50, 100, 600],
    labels=["low", "mid", "high", "premium"]
)


# =========================
# Analysis
# =========================

# Survival rate by Title
title_survival = df.groupby("Title")["Survived"].mean()

print("===== Survival Rate by Title =====")
print(title_survival)


# Survival rate by FamilySize
family_survival = df.groupby("FamilySize")["Survived"].mean()

print("\n===== Survival Rate by Family Size =====")
print(family_survival)


# Survival rate by FareCategory
fare_survival = df.groupby(
    "FareCategory",
    observed=False
)["Survived"].mean()

print("\n===== Survival Rate by Fare Category =====")
print(fare_survival)


# =========================
# Save Cleaned Dataset
# =========================

df.to_csv("titanic_clean.csv", index=False)

print("\nCleaned dataset saved as 'titanic_clean.csv'")


# =========================
# 5-Bullet Summary
# =========================

female_rate = df[df["Sex"] == "female"]["Survived"].mean()

male_rate = df[df["Sex"] == "male"]["Survived"].mean()

print("\n===== SUMMARY =====")

print(
    f"1. Females had a survival rate of "
    f"{female_rate:.2f}, while males had {male_rate:.2f}."
)

print(
    "2. Passengers paying higher fares generally "
    "had better survival chances."
)

print(
    "3. Small families survived more often than "
    "very large families."
)

print(
    "4. Many passengers travelling alone "
    "did not survive."
)

print(
    "5. Titles like 'Mrs' and 'Miss' had "
    "higher survival rates than 'Mr'."
)
```
