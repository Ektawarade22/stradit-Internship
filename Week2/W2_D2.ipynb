```python
#TASK 1  write_lines(path, lines)
def write_lines(path, lines):
    with open(path, "w") as file:
        for line in lines:
            file.write(line + "\n")


# Data to write into the file
data = ["Apple", "Banana", "Mango"]

# Calling the function
write_lines("file1.txt", data)

print("Data written successfully!")

```

```python
#TASK 2 count non empty lines
def count_lines(path):
    try:
        with open("file.txt", "r") as file:
            lines = file.readlines()

            count = 0

            for line in lines:
                if line.strip() != "":
                    count += 1

            return count

    except FileNotFoundError:
        return 0


# Function call
result = count_lines("file1.txt")

print(result)
```

```python
#TASK 3 Count total words
def word_count(path):
    with open("File1.txt", "r") as file:
        text = file.read()
        words = text.split()
        return len(words)


# Function call
result = word_count("file1.txt")

print(result)

```

```python
#TASK 4 return last n lines
def tail(path, n):
    with open(path, "r") as file:
        lines = file.readlines()
        return lines[-n:]


# Function call
result = tail("file.txt", 2)

print(result)
```

```python
#TASK 2 Adding logs
from datetime import datetime

def log(msg):
    timestamp = datetime.now().isoformat()

    with open("app.log", "a") as file:
        file.write(f"{timestamp} {msg}\n")


# Function calls
log("Program started")
log("User logged in")

print("Logs added successfully!")
```

```python
#TASK 3 Replace the content
def replace_in_file(path, old, new):

    with open(path, "r") as file:
        text = file.read()

    text = text.replace(old, new)

    with open(path, "w") as file:
        file.write(text)


# Function call
replace_in_file("file.txt", "Functions", "Array")

print("Text replaced successfully!")
```

```python
# Task 1 
import csv
data = [
    ["name", "age", "city"],
    ["Rahul", 22, "Mumbai"],
    ["Anjali", 20, "Delhi"],
    ["Amit", 23, "Nagpur"],
    ["Sneha", 24, "Nashik"]
]

with open("students.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("CSV file created successfully!")

print("\n Using csv.reader:\n")

with open("students.csv", "r") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)

print("\n Using csv.DictReader:\n")

with open("students.csv", "r") as file:
    dict_reader = csv.DictReader(file)

    for row in dict_reader:
        print(row)
```

```python
#task 2
import csv

# open people.csv for reading
with open("people.csv", "r") as infile:

    # open adults.csv for writing
    with open("adults.csv", "w", newline="") as outfile:

        reader = csv.DictReader(infile)

        fieldnames = ["name", "age", "city"]

        writer = csv.DictWriter(outfile, fieldnames=fieldnames)

        # write header
        writer.writeheader()

        # check each row
        for row in reader:

            # keep only age >= 30
            if int(row["age"]) >= 30:
                writer.writerow(row)

print("adults.csv created")
```

```python
#task 4
import csv

def csv_to_dict_list(path):

    data = []

    with open(path, "r") as file:

        reader = csv.DictReader(file)

        for row in reader:
            data.append(row)

    return data


result = csv_to_dict_list("students.csv")

print(result)
```

```python
#task 5
import csv

def dict_list_to_csv(records, path):

    with open(path, "w", newline="") as file:

        fieldnames = records[0].keys()

        writer = csv.DictWriter(file, fieldnames=fieldnames)

        writer.writeheader()

        writer.writerows(records)


students = [
    {"name": "Ekta", "score": 80},
    {"name": "Rahul", "score": 70},
    {"name": "Sneha", "score": 90}
]

dict_list_to_csv(students, "new_students.csv")

print("new_students.csv created")
```

```python
#task 1
def safe_int(s, default=0):

    try:
        return int(s)

    except ValueError:
        return default


print(safe_int("25"))
print(safe_int("abc"))
print(safe_int("abc", -1))
```

```python
#task 2
def open_file():

    while True:

        filename = input("Enter file name: ")

        try:
            with open(filename, "r") as file:

                content = file.read()

                print("\n File opened successfully")
                print(content)

                break

        except FileNotFoundError:
            print("File not found. Try again.")


open_file()
```

```python
#task 3
def divide(a, b):

    try:
        return a / b

    except TypeError:
        print("Invalid data type")
        return None

    except ZeroDivisionError:
        print("Cannot divide by zero")
        return None


print(divide(10, 2))
print(divide(10, 0))
print(divide(10, "a"))
```

```python
#task 4
def read_numbers(path):

    numbers = []
    skipped = 0

    with open(path, "r") as file:

        for line in file:

            line = line.strip()

            try:
                number = float(line)
                numbers.append(number)

            except ValueError:
                skipped += 1

    print("Valid numbers:", numbers)
    print("Skipped lines:", skipped)


read_numbers("numbers.txt")
```

```python
#task 5
class EmptyFileError(Exception):
    pass


def read_file(path):

    with open(path, "r") as file:

        content = file.read()

        if content.strip() == "":
            raise EmptyFileError("File is empty")

        return content


try:
    data = read_file("numbers.txt")
    print(data)

except EmptyFileError as e:
    print(e)
```

```python
import csv

try:

    # open input file
    with open("students.csv", "r") as infile:

        reader = csv.DictReader(infile)

        # open output file
        with open("students_with_grades.csv", "w", newline="") as outfile:

            fieldnames = [
                "name",
                "math",
                "science",
                "english",
                "total",
                "average"
            ]

            writer = csv.DictWriter(outfile, fieldnames=fieldnames)

            writer.writeheader()

            # process each student
            for row_number, row in enumerate(reader, start=1):

                total = 0

                subjects = ["math", "science", "english"]

                for subject in subjects:

                    try:
                        marks = int(row[subject])

                    except ValueError:

                        print(
                            f"Warning: Invalid value in row {row_number}, column '{subject}'. Using 0."
                        )

                        marks = 0

                    total += marks

                average = total / 3

                # add new columns
                row["total"] = total
                row["average"] = average

                writer.writerow(row)

    print("\nstudents_with_grades.csv created successfully")


except FileNotFoundError:

    print("Error: students.csv file not found")
```
