```python
#1 Circle Calculator

pi = 3.14159

# Function to calculate area of a circle
def circle_area(radius):
    return pi * radius * radius

# Function to calculate circumference of a circle
def circle_c(radius):
    return 2 * pi * radius


radii = [1, 5, 10]
for r in radii:
    area = circle_area(r)
    circumference = circle_c(r)

    print(f"Radius: {r}")
    print(f"Area: {area:.2f}")
    print(f"Circumference: {circumference:.2f}")
    print("_" * 30)

```

```python
#2 Palindrome Checker
def is_palindrome(text):
  text=text.lower()
  return text==text[::-1]

words=["madam","racecar","hello","civic"]
for word in words:
    print(f"{words}->{is_palindrome(word)}")
```

```python
#3 Password Generator
import random
import string

def gen_password(len=12):
  char=string.ascii_letters+string.digits+string.punctuation
  password=""
  for i in range(len):
    password+=random.choice(char)
  return password

for i in range(5):
  print(gen_password())
```

```python
#4 Sentence analyser
def analyse_sentence(text):
    words = text.split()

    # Count words
    word_count = len(words)

    # Count characters excluding spaces
    char_count = len(text.replace(" ", ""))

    vowels = "aeiouAEIOU"
    vowel_count = 0
    consonant_count = 0

    # Count vowels and consonants
    for char in text:
        if char.isalpha():
            if char in vowels:
                vowel_count += 1
            else:
                consonant_count += 1

    # Find longest and shortest word
    longest_word = max(words, key=len)
    shortest_word = min(words, key=len)

    result = {
        "word_count": word_count,
        "char_count": char_count,
        "vowel_count": vowel_count,
        "consonant_count": consonant_count,
        "longest_word": longest_word,
        "shortest_word": shortest_word
    }

    return result

text = "Python is easy language as compared to any other programming language"
output = analyse_sentence(text)
print(output)
```

```python
#5 Converter Toolkit
# Functions
def km_to_miles(km):
    return km * 0.621371

def kg_to_pounds(kg):
    return kg * 2.20462

def litres_to_gallons(l):
    return l * 0.264172

while True:

    print("\n Converter Toolkit ")
    print("1. KM to Miles")
    print("2. KG to Pounds")
    print("3. Litres to Gallons")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        km = float(input("Enter kilometers: "))
        print("Miles:", km_to_miles(km))

    elif choice == "2":
        kg = float(input("Enter kilograms: "))
        print("Pounds:", kg_to_pounds(kg))

    elif choice == "3":
        litres = float(input("Enter litres: "))
        print("Gallons:", litres_to_gallons(litres))

    elif choice == "4":
        print("Exit")
        break

    else:
        print("Invalid choice")
```

```python
#6 Build a mini report card system

# Function
def get_student_info():
    name = input("Enter student name: ")
    marks = []

    for i in range(1, 6):
        mark = float(input(f"Enter marks for subject {i}: "))
        marks.append(mark)

    return name, marks


def calculate_stats(marks):

    total = sum(marks)
    average = total / len(marks)
    highest = max(marks)
    lowest = min(marks)

    return total, average, highest, lowest


def determine_grade(average):

    if average >= 90:
        return "A+"

    elif average >= 80:
        return "A"

    elif average >= 70:
        return "B"

    elif average >= 60:
        return "C"

    elif average >= 50:
        return "D"

    else:
        return "F"


# Function to print report card
def print_report(name, marks, stats, grade):

    total, average, highest, lowest = stats

    print("\n REPORT CARD ")
    print(f"Student Name : {name}")
    print(f"Marks : {marks}")

    print(f"Total Marks : {total}")
    print(f"Average     : {average:.2f}")

    print(f"Highest Mark : {highest}")
    print(f"Lowest Mark  : {lowest}")

    print(f"Grade : {grade}")

    print("=" * 30)


name, marks = get_student_info()

stats = calculate_stats(marks)

grade = determine_grade(stats[1])

print_report(name, marks, stats, grade)
```
