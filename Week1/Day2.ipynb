```python
#2
n1=int(input("Enter First number:"))
n2=int(input("Enter second number :"))
#Result
print(f"sum={n1+n2}")
print(f"diff={n1-n2}")
print(f"product={n1*n2}")
print(f"Division={n1/n2}")
print(f"Floor division={n1//n2}")
print(f"Modulus={n1%n2}")
```

```python
#5
word=input("Enter any word")
print(f"Length of word={len(word)}")
print(f"Uppercase={word.upper()}")
print(f"Lowercase={word.lower()}")
print(f"FirstChar={word[0]}")
print(f"LastChar={word[-1]}")
print(f"Reversed={word[::-1]}")


```

```python
#6
price=float(input("Enter the item Price:"))
quantity=int(input("Enter the quantity:"))
subtotal=price*quantity
tax=0.18*subtotal
total=subtotal+tax
#Formated bill
print(f"Price:{price}")
print(f"Quantity:{quantity}")
print(f"Subtotal:{subtotal}")
print(f"Tax:{tax}")
print(f"Total:{total}")

```
