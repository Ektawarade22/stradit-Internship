```python
#1 Age category classifier
age = int(input("Enter the age: "))
if age >= 0 and age <= 12:
    print("category = Child")
elif age >= 13 and age <= 17:
    print("category = Teenager")
elif age >= 18 and age <= 24:
    print("category = Young Adult")
elif age >= 25 and age <= 59:
    print("category = Adult")
elif age >= 60 and age <= 150:
    print("category = Senior")
else:
    print("Invalid age")
```

```python
#2 Simple calculator
n1 = float(input("Enter first number:"))
n2 = float(input("Enter second number:"))
op= input("Enter operator:")
if op== "+": #addition
    print(n1+n2) 
elif op== "-": #substraction
    print(n1-n2)
elif op== "*": #multiplication
    print(n1*n2)
elif op== "/": #division
    if n2 == 0:
        print("Cannot divide by zero")
    else:
        print(n1/n2)
elif op== "//": #floor division
    if n2 == 0:
        print("Cannot divide by zero")
    else:
        print(n1//n2)
elif op== "%": #modulus
    if n2 == 0:
        print("Cannot divide by zero")
    else:
        print(n1%n2)
elif op=="**": #power
    print(n1**n2)
else:
    print("Invalid Operator")
```

```python
#3 Multiplication table generator
num=int(input("Enter the number:"))
for i in range(1, 13):
    print(f"{num} x {i} = {num * i}")
```

```python
#3(type2) using while loop
number = int(input("Enter a number: "))
start = int(input("Enter start range: "))
end = int(input("Enter end range: "))
i = start
while i <= end:
    print(number, "x", i, "=", number * i)
    i += 1
```

```python
#4 Number pattern printer
#Pattern A (right traingles of numbers) (using for loop)
for i in range(1,6):
    for j in range(1,i + 1):
        print(j, end=" ")
    print()
```

```python
#Pattern B (inverted right traingles of numbers) 
for i in range(5,0,-1):
    for j in range(i,0,-1):
        print(j, end=" ")
    print()
```

```python
#5 Digit sum calculator
num = int(input("Enter a number: "))
sum=0
product=1
while num > 0:
    digit = num % 10
    sum += digit
    product *= digit
    num = num // 10
print("Sum =", sum)
print("Product =", product)
```

```python
#6 Password strength checker
password = input("Enter password: ")
length = len(password)
has_digit = False
has_upper = False
for ch in password:
    if ch.isdigit():
        has_digit = True
    if ch.isupper():
        has_upper = True

if length < 6:  # Length check
    status = "Weak"
elif length <= 9:
    status = "Medium"
else:
    status = "Strong"

print("Length Status:", status)
print("Contains Digit:", has_digit)
print("Contains Uppercase:", has_upper)

if length >= 10 and has_digit and has_upper:  #rating
    rating = "Strong"
elif length >= 6 and (has_digit or has_upper):
    rating = "Medium"
else:
    rating = "Weak"

print("Password Strength:", rating)
```

```python
# Daimond pattern
rows = int(input("Enter number of rows: "))
for i in range(1, rows + 1):  #upper part
    for j in range(rows - i):  #for space
        print(" ", end="")
    for j in range(2 * i - 1): #for printing stars
        print("*", end="")
    print()

for i in range(rows - 1, 0, -1):  #lower part
    for j in range(rows - i):
        print(" ", end="")
    for j in range(2 * i - 1):
        print("*", end="")
    print()
```
